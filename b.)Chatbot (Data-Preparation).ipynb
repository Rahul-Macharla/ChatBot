```python
import pandas as pd
import numpy as np
```

```python
import re
import nltk
import string
import nlp_utils as nu
import contractions
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize,sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
```

```python
from sklearn.metrics.pairwise import cosine_distances
```

```python
df=pd.read_csv('NormalisedDataFrame.csv',names=('Context','Response'))
```

```python
df.head()
```

```python
df = df.astype(str) # type convertion
```

```python
df['Context']=df['Context'].str.replace("_"," ")
df['Context']=df['Context'].str.replace("_"," ")
df['Context']=df['Context'].str.replace("eoa"," ")
df['Context']=df['Context'].str.replace("eaot"," ")
df['Context']=df['Context'].str.replace("eaut"," ")
df['Context']=df['Context'].str.replace("eau"," ")
df['Context']=df['Context'].str.replace("eou"," ")
```

```python
df['Response']=df['Response'].str.replace("_"," ")
df['Response']=df['Response'].str.replace("_"," ")
df['Response']=df['Response'].str.replace("eoa"," ")
df['Response']=df['Response'].str.replace("eaot"," ")
df['Response']=df['Response'].str.replace("eaut"," ")
df['Response']=df['Response'].str.replace("eau"," ")
df['Response']=df['Response'].str.replace("eou"," ")
```

```python
df=df.dropna()
```

```python
df
```

```python
# Text Normalization on Context.
df['Context'] = df['Context'].apply(nu.lemmatization_sentence)
# Word Embedding - TF-IDF
tfidf = TfidfVectorizer()
factors = tfidf.fit_transform(df['Context']).toarray()
```

```python
df
```

# Saving

```python
dfw = pd.DataFrame(df)
  
# saving the dataframe 
dfw.to_csv('Tf-Idf.csv', header=False, index=False) 
```
